Create Dataset

```python
import pandas as pd
```

```python
StringStart = '{"label":"SOFT-SKILL","pattern":['
StringEnd = ']}'
extract = pd.read_csv('data/softSkills.csv', encoding='ISO-8859-1')
```

```python
extract
```

```python
Array = []
for e in extract['soft_skill']:
    Array.append(e.split(' '))
    #print(e.split(' '))
    
Array2 = []
for line in Array:
    Array2.append([word for word in line if word != ''])
Array2
```

```python
dataset = []
for skill in Array2:
    fullSTR = ""
    for word in skill:
        string ='{"LOWER":"'+word+'"},'
        fullSTR=fullSTR+string
    fullSTR = fullSTR[:-1]
    dataset.append(StringStart+fullSTR+StringEnd)
```

```python
dataset
```

```python
with open ("newDataSet.txt", 'w', encoding="utf-8") as f:
    for line in dataset:
        f.write(line +"\n")
```
